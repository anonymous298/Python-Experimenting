### Handling Imbalanced Dataset

```python
sns.scatterplot(data=df, x='age', y='bmi', hue='diabetes')
```

```python
# Splitting Data into Dependent and Independent features

X, y = df.drop(['diabetes'], axis=1), df['diabetes']
```

##### Using Undersampling Technique

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=0)
X_undersampled, y_undersampled = rus.fit_resample(X, y)
```

```python
# Acutal output feature value counts
y.value_counts()
```

```python
# Undersampled output feature value counts
y_undersampled.value_counts()
```

```python
sns.scatterplot(x=X_undersampled['age'], y=X_undersampled['bmi'], hue=y_undersampled)
```

#### Using Oversampled Technique 

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=0)
X_oversampled, y_oversampled = ros.fit_resample(X, y)
```

```python
sns.scatterplot(x=X_oversampled['age'], y=X_oversampled['bmi'], hue=y_oversampled)
```

```python
# Acutal output feature value counts
y.value_counts()
```

```python
# Undersampled output feature value counts
y_oversampled.value_counts()
```

### Creating a preprocessing Pipeline

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
numerical_columns = df.select_dtypes('number').columns.drop(['diabetes'])
categorical_columns = df.select_dtypes('object').columns
```

```python
numerical_pipeline = Pipeline(
    [
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaling', StandardScaler())
    ]
)

categorical_pipeline = Pipeline(
    [
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoding', OneHotEncoder())
    ]
)
```

```python
preprocessor = ColumnTransformer(
    [
        ('num_prerpocessor', numerical_pipeline, numerical_columns),
        ('categorical_preprocessor', categorical_pipeline, categorical_columns)
    ]
)
```

```python
X_transformed = preprocessor.fit_transform(X).toarray()
```

```python
X_transformed.shape
```

### Using SMOTE Technique

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=0)
X_resampled, y_resampled = sm.fit_resample(X_transformed, y)
```

```python
sns.scatterplot(x=X_resampled[:, 0], y=X_resampled[:, 1], hue=y_resampled)
```

```python
# Acutal output feature value counts
y.value_counts()
```

```python
# SMOTE output feature value counts
y_resampled.value_counts()
```
